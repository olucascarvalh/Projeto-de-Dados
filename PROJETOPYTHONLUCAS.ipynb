```python
# PASSO A PASSO - Você é um analista de dados de uma empresa muito grande que vende
#no varejo através de lojas, seu dever é gerar análises em cima dos dados que tem
# 
#PASSO 01: Importar a base de dados
#PASSO 02: analisar a base de dados
#PASSO 03: mostrar a parte tras da base de dados 
#PASSO 04: corrigir as cagadas da base de dados
#PASSO 05:analisar possibilidade de criação de dados

#PASSO 06: PONTO COMPARATIVO:
# LOJAS COM MAIOR  FATURAMENTO - PRODUTOS MAIS VENDIDOS - 
#PASSO 07: gerar dashboard 
#PASSO 08: importar para BI novo banco de dados com as modificações criadas
#PASSO 09: criar dashboard


#PASSO 01: importar a base de dados
import pandas as pd

# USANDO 'latin1' se o arquivo estiver nessa codificação
bancod = pd.read_csv("Vendas - Dez.csv", encoding='latin1')
display(bancod)

#PASSO 02: analisar a base de dados
```

```python
#PASSO 03: mostrar banco de dados por trás para ver dados quebrados

display(bancod.info())
```

```python
import pandas as pd

# Contar a quantidade de cada tipo de loja
ranking_lojas = bancod['ID Loja'].value_counts().reset_index()

# Renomear as colunas para algo mais significativo
ranking_lojas.columns = ['ID Loja', 'Quantidade']

print(ranking_lojas)
```

```python
import pandas as pd

# Calcular o valor total das vendas por loja
valor_vendas_lojas = bancod.groupby('ID Loja')['Valor Final'].sum().reset_index()

# Adicionar o ranking com base no valor total das vendas
valor_vendas_lojas = valor_vendas_lojas.sort_values(by='Valor Final', ascending=False)
valor_vendas_lojas['Ranking'] = range(1, len(valor_vendas_lojas) + 1)

display(valor_vendas_lojas)
```

```python
import plotly.express as px
import pandas as pd

# Criar o DataFrame com o valor total das vendas por loja
valor_vendas_lojas = bancod.groupby('ID Loja')['Valor Final'].sum().reset_index()

# Adicionar o ranking com base no valor total das vendas
valor_vendas_lojas = valor_vendas_lojas.sort_values(by='Valor Final', ascending=False)
valor_vendas_lojas['Ranking'] = range(1, len(valor_vendas_lojas) + 1)

# Criar um gráfico de barras do ranking de lojas por valor de vendas, com variação de cor por ID Loja
grafico_ranking = px.bar(valor_vendas_lojas, x='ID Loja', y='Valor Final', color='ID Loja',
                         title='Total de Vendas por Lojas', labels={'Valor Final': 'Valor Final'})

# Exibir o gráfico
grafico_ranking.show()

```

```python
import pandas as pd
import plotly.express as px

# Calcular a quantidade total de vendas para cada tipo de produto
quantidade_vendas = bancod.groupby('Produto')['Quantidade'].sum().reset_index()

# Adicionar a coluna de ranking com base na quantidade de vendas
quantidade_vendas = quantidade_vendas.sort_values(by='Quantidade', ascending=False)
quantidade_vendas['Ranking'] = range(1, len(quantidade_vendas) + 1)

# Exibir a quantidade total de vendas para cada produto com ranking
print(quantidade_vendas)
```

```python
# Criar um gráfico de barras do ranking de lojas por valor de vendas, com variação de cor por ID Loja
VendasProduto = px.bar(quantidade_vendas, x='Produto', y='Quantidade', color='Produto',
                         title='Produtos + Vendidos', labels={'Quantidade': 'Quantidade'})

# Exibir o gráfico
VendasProduto.show()
```
